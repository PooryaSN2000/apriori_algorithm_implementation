**Apriori Implementation**

**0.** Library

```python
from collections import Counter
import random
import itertools
```

**1.** Build list item

```python
print("all item we use to bulid random dataset :")
item_list = ['A','B','C','D','E','F','G','H','I','J','K','L','M']
print(item_list)
dataset = []
for i in range (0,200):
    n = random.randint(1,7)
    dataset.append(['T{}'.format((i+1)),sorted(random.sample(item_list, k=n))])
    
#all database
for j in range (0,200):
    print(dataset[j])
print("data we select :")
#sampling method:
samplingDataset = random.choices(dataset,k=20)
datasetMain = dataset
dataset = samplingDataset
```

**2.** Make Dataset

```python
def MakeDataset(dataset):
    itemUsed = []
    for i in dataset:
        for j in i[1]:
            if(j not in itemUsed):
                itemUsed.append(j)
    itemUsed = sorted(itemUsed)

    # We specify the sup(min) value :
    sup = float(input("Enter SUP(min) Persentage like 0.1 as 10% of transaction: "))
    sup = float(sup*len(dataset))
    print("minSUP :")
    print(sup)
    print("an item we use in our dataset :")
    print(itemUsed)
    print("our data we use :")
    for item in dataset:
         print(item)
    return itemUsed , sup
```

```python
itemUsed , sup = MakeDataset(dataset)
```

**3.** Create Candidate 1 (C1) and Frequent 1 (L1)

```python
def MakeCandidate(itemUsed,dataset):
    Candidate = Counter()
    for i in itemUsed:
        for j in dataset:
            if(i in j[1]):
                Candidate[i]+=1
    print("Candidate C1:")
    for i in Candidate:
        print(str(list(i))+": "+str(Candidate[i]))
    return Candidate
```

```python
Candidate = MakeCandidate(itemUsed,dataset)
```

```python
def MakeFrequent(Candidate,sup):
    Frequent = Counter()
    for i in Candidate:
        if(Candidate[i] >= sup):
            Frequent[frozenset([i])]+=Candidate[i]
    print("Frequent L1:")
    for i in Frequent:
        print(str(list(i))+": "+str(Frequent[i]))
        
    return Frequent
```

```python
Frequent = MakeFrequent(Candidate,sup)
```

**4.** CandidateMaker && FrequentMaker

```python
def CandidateMaker(nCandidate,dataset,count):
    Candidate = Counter()
    for i in nCandidate:
        Candidate[i] = 0
        for j in dataset:
            if(i.issubset(set(j[1]))):
                Candidate[i]+=1
    #print part 
    print("C"+str(count)+":")
    if(len(Candidate) == 0):
        print("There is nothing left") 
    else:
        for i in Candidate:
            print(str(list(i))+": "+str(Candidate[i]))
    return Candidate
```

```python
def FrequentMaker(Candidate,sup,count):
    Frequent = Counter()
    for i in Candidate:
        if(Candidate[i] >= sup):
            Frequent[i]+=Candidate[i]
    #print part        
    print("L"+str(count)+":")
    if(len(Frequent) == 0):
        print("There is nothing left") 
    else:
        for i in Frequent:
            print(str(list(i))+": "+str(Frequent[i]))
    return Frequent
```

**5.** Main Loop

```python
def findsubsets(s, n):
    return list(map(set, itertools.combinations(s, n)))
def useless(pFrequent,merge,i,j):
    temp = merge[i].union(merge[j])
    allSubSet = findsubsets(temp,len(temp)-1)
    listpFre = [set(x) for x in pFrequent]
    result =  all(elem in listpFre for elem in allSubSet)
    return result
```

```python
def Loop(Frequent,dataset,sup):
    #previousFrequent ==> pFrequent
    pFrequent = Frequent
    postion = 1
    for count in range (2,8):
        newCandidate = set()
        merge = list(Frequent)
        for i in range(0,len(merge)):
            for j in range(i+1,len(merge)):
                if(len(merge[i].union(merge[j])) == count):
                    if(useless(pFrequent,merge,i,j)):
                        newCandidate.add(merge[i].union(merge[j]))

        newCandidate = list(newCandidate) 
        #new Candidate lsit MAKER
        Candidate = CandidateMaker(newCandidate,dataset,count)
        #new Frequent lsit MAKER
        Frequent = FrequentMaker(Candidate,sup,count)
        if(len(Frequent) == 0):
            break
        pFrequent = Frequent
        postion = count 
    return pFrequent,postion
```

```python
pFrequent , postion = Loop(Frequent,dataset,sup)
```

**6.** Check ANSWER with sample database

```python
def Answer(postion,pFrequent):
    print("Answer: ")
    print("L"+str(postion)+":")
    for i in pFrequent:
        print(str(list(i))+": "+str(pFrequent[i]))
        
    return pFrequent,postion
```

```python
FrequentSampleDatabase , PostionSampleDatabase = Answer(postion,pFrequent)
```

**7.** Check ANSWER with main database

```python
itemUsed , sup = MakeDataset(datasetMain)
Candidate = MakeCandidate(itemUsed,datasetMain)
Frequent = MakeFrequent(Candidate,sup)
pFrequent , postion = Loop(Frequent,datasetMain,sup)
FrequentMainDatabase , PostionMainDatabase = Answer(postion,pFrequent)
```
